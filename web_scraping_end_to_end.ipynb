```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib.parse
```

```python
website = 'https://www.laptopsdirect.co.uk/ct/laptops-and-netbooks/laptops?fts=laptop'
```

```python
 response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
results=soup.find_all('div',{'class':'OfferBox'})
```

```python
product_name=[]
product_price=[]
product_info=[]

for result in results:
  try:
    product_name.append(result.find('a' ,{'class': 'offerboxtitle'}).get_text())
  except:
    product_name.append('n/a')

  try:
    product_price.append(result.find('span' ,{'class': 'offerprice'}).get_text()).strip()
  except:
    product_price.append('n/a')

  try:
    product_info.append(result.find('div' ,{'class': 'productInfo'}).get_text()).strip()
  except:
    product_info.append('n/a')
```

```python
product_name=[]
product_price=[]
product_info=[]
relative_urls=[]
combined_urls=[]

for result in results:
    product_name.append(result.find('a' ,{'class': 'offerboxtitle'}).get_text())


    product_price.append(result.find('span' ,{'class': 'offerprice'}).get_text())

    product_info.append(result.find('div' ,{'class': 'productInfo'}).get_text())

    relative_urls.append(result.find('a',{'class' : 'offerboxtitle'}).get('href'))

for link in relative_url:
     combined_urls.append(urllib.parse.urljoin(root_url,link))
```
